Question_5

```python
q5=df.iloc[1155:1587]
q5=q5[['Births by women aged 15 to 19 (thousands)','Year','Region, subregion, country or area *']].set_index('Region, subregion, country or area *')
```

```python
high_income_countries= q5.loc["High-income countries",['Year', 'Births by women aged 15 to 19 (thousands)']]
No_income_group_available =q5.loc["No income group available",['Year', 'Births by women aged 15 to 19 (thousands)']]
Low_income_countries =q5.loc["Low-income countries",['Year', 'Births by women aged 15 to 19 (thousands)']]
Lower_middle_income_countries =q5.loc["Lower-middle-income countries",['Year', 'Births by women aged 15 to 19 (thousands)']]
Upper_middle_income_countries =q5.loc["Upper-middle-income countries",['Year', 'Births by women aged 15 to 19 (thousands)']]
Middle_income_countries =q5.loc["Middle-income countries",['Year', 'Births by women aged 15 to 19 (thousands)']]
```

```python
plt.plot( high_income_countries['Year'],high_income_countries['Births by women aged 15 to 19 (thousands)'], label = "high_income_countries");

plt.plot( No_income_group_available['Year'],No_income_group_available['Births by women aged 15 to 19 (thousands)'], label = "No_income_group_available");

plt.plot( Low_income_countries['Year'],Low_income_countries['Births by women aged 15 to 19 (thousands)'], label = "high_income_countries");

plt.plot( Lower_middle_income_countries['Year'],Lower_middle_income_countries['Births by women aged 15 to 19 (thousands)'], label = "Lower_middle_income_countries");


plt.plot( Upper_middle_income_countries['Year'],Upper_middle_income_countries['Births by women aged 15 to 19 (thousands)'], label = "Upper_middle_income_countries");


plt.plot( Middle_income_countries['Year'],Middle_income_countries['Births by women aged 15 to 19 (thousands)'], label = "Middle_income_countries");
plt.xlabel('Year') 
plt.ylabel('Births by women aged 15 to 19 (thousands)') 
plt.title('Births by women between 15 and 19 years old based on income level and year') 
plt.legend(bbox_to_anchor=(1.1, 1));






```

Question_8


```python
q8=df[df['Region, subregion, country or area *']=='WORLD']
Male_Life_15=q8[['Male Life Expectancy at Age 15 (years)']].sum().values[0]
Female_Life_15=  q8[['Female Life Expectancy at Age 15 (years)' ]].sum().values[0]
Male_Life_65 =  q8[['Male Life Expectancy at Age 65 (years)']].sum().values[0]
Female_Life_65 =   q8[['Female Life Expectancy at Age 65 (years)']].sum().values[0]
Male_Life_80 =q8[['Male Life Expectancy at Age 80 (years)']].sum().values[0]
Female_Life_80 = q8[['Female Life Expectancy at Age 80 (years)']].sum().values[0]
ages=['15','65','80']
male=[Male_Life_15,Male_Life_65,Male_Life_80 ]
female=[Female_Life_15,Female_Life_65,Female_Life_80 ]
```

```python
plt.bar(ages,male, label = 'male')

plt.bar(ages,female,bottom=male, label='female')

plt.legend();
```
