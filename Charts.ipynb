```python
q8=df[df['Region, subregion, country or area *']=='WORLD']
Male_Life_15=q8[['Male Life Expectancy at Age 15 (years)']].sum().values[0]
Female_Life_15=  q8[['Female Life Expectancy at Age 15 (years)' ]].sum().values[0]
Male_Life_65 =  q8[['Male Life Expectancy at Age 65 (years)']].sum().values[0]
Female_Life_65 =   q8[['Female Life Expectancy at Age 65 (years)']].sum().values[0]
Male_Life_80 =q8[['Male Life Expectancy at Age 80 (years)']].sum().values[0]
Female_Life_80 = q8[['Female Life Expectancy at Age 80 (years)']].sum().values[0]
ages=['15','65','80']
male=[Male_Life_15,Male_Life_65,Male_Life_80 ]
female=[Female_Life_15,Female_Life_65,Female_Life_80 ]
```
